Task 1: Plot X-Axis as "Country", Y-Axis as "Confirmed"


```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_csv("/content/covid19.csv")
df.head()
```

```
%matplotlib inline
#plt.figure(figsize=(15,5))
plt.xlabel('Country')
plt.ylabel('Confirmed')
plt.plot(df['Country'],df['Confirmed'])
#plt.xticks(rotation=90)
plt.show()
```

Task 2: Plot X-Axis as "Country", Y-Axis as "Deaths"


```
%matplotlib inline
#plt.figure(figsize=(15,5))
plt.xlabel('Country')
plt.ylabel('Deaths')
plt.plot(df['Country'],df['Deaths'])
#plt.xticks(rotation=90)
plt.show()
```

Task 13: Annotate "Confirmed" High Pick as "ConPick"


```
%matplotlib inline
plt.figure(figsize=(15,5))
plt.xlabel('Country')
plt.ylabel('Confirmed-Death')
plt.title("JINENDRA_MEHTA_180570107070")
plt.grid()
pick_confirmed=max(df['Confirmed'])
plt.annotate(xy=[df[df.Confirmed==df.Confirmed.max()].Country,pick_confirmed],s='Conpeak')
plt.plot(df['Country'],df['Confirmed'],"k-->",label="Confirmed")
plt.plot(df['Country'],df['Deaths'],"r:o",label="Deaths")
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

Task 14: Annotate "Death" High Pick as "DeathPick"


```
%matplotlib inline
plt.figure(figsize=(15,5))
plt.xlabel('Country')
plt.ylabel('Confirmed-Death')
plt.title("JINENDRA_MEHTA_180570107070")
plt.grid()
pick_confirmed=max(df['Confirmed'])
pick_daths=max(df["Deaths"])
plt.annotate(xy=[df[df.Confirmed==df.Confirmed.max()].Country,pick_confirmed],s='Conpeak')
plt.annotate(xy=[df[df.Deaths==df.Deaths.max()].Country,pick_daths],s='DeathPeak')
plt.plot(df['Country'],df['Confirmed'],"k-->",label="Confirmed")
plt.plot(df['Country'],df['Deaths'],"r:o",label="Deaths")
plt.xticks(rotation=90)
plt.legend()
plt.show()
```
