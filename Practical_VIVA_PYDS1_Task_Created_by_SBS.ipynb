```
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("/content/covid19.csv")
```

```
df.columns
```

```
plt.figure(figsize=(10,5))
plt.xlabel('WHOR')
plt.ylabel('New cases')
plt.plot(df['WHOR'],df['New cases'])
plt.show()
```

```
plt.figure(figsize=(10,5))
plt.xlabel('WHOR')
plt.ylabel('New deaths')
plt.plot(df['WHOR'],df['New deaths'])
plt.show()
```

```
plt.figure(figsize=(10,5))
plt.xlabel('WHOR')
plt.ylabel('New cases/New deaths')
plt.plot(df['WHOR'],df['New cases'])
plt.plot(df['WHOR'],df['New deaths'])
plt.show()
```

```
plt.figure(figsize=(10,5))
plt.xlabel('WHOR')
plt.ylabel('New cases/New deaths')
plt.plot(df['WHOR'],df['New cases'])
plt.plot(df['WHOR'],df['New deaths'])
plt.xticks(rotation=90)
plt.show()
```

```
plt.figure(figsize=(10,5))
plt.xlabel('WHOR')
plt.ylabel('New cases-deaths')
plt.plot(df['WHOR'],df['New cases'])
plt.plot(df['WHOR'],df['New deaths'])
plt.xticks(rotation=90)
plt.show()
```

```
plt.figure(figsize=(10,5))
plt.xlabel('WHOR')
plt.ylabel('New cases-deaths')
plt.plot(df['WHOR'],df['New cases'],label='New cases')
plt.plot(df['WHOR'],df['New deaths'],label='New deaths')
plt.legend()
plt.xticks(rotation=90)
plt.show()
```

```
plt.figure(figsize=(10,5))
plt.title("JINENDRA_MEHTA_180570107070")
plt.xlabel('WHOR')
plt.ylabel('New cases-deaths')
plt.plot(df['WHOR'],df['New cases'])
plt.plot(df['WHOR'],df['New deaths'])
plt.xticks(rotation=90)
plt.show()
```

```
plt.figure(figsize=(10,5))
plt.title("JINENDRA_MEHTA_180570107070")
plt.grid()
plt.xlabel('WHOR')
plt.ylabel('New cases-deaths')
plt.plot(df['WHOR'],df['New cases'])
plt.plot(df['WHOR'],df['New deaths'])
plt.xticks(rotation=90)
plt.show()
```

```
plt.figure(figsize=(10,5))
plt.title("JINENDRA_MEHTA_180570107070")
plt.grid()
plt.xlabel('WHOR')
plt.ylabel('New cases-deaths')
plt.plot(df['WHOR'],df['New cases'],'k--')
plt.plot(df['WHOR'],df['New deaths'])
plt.xticks(rotation=90)
plt.show()
```

```
plt.figure(figsize=(10,5))
plt.title("JINENDRA_MEHTA_180570107070")
plt.grid()
plt.xlabel('WHOR')
plt.ylabel('New cases-deaths')
plt.plot(df['WHOR'],df['New cases'],'k--')
plt.plot(df['WHOR'],df['New deaths'],'r:')
plt.xticks(rotation=90)
plt.show()
```

```
plt.figure(figsize=(10,5))
plt.title("JINENDRA_MEHTA_180570107070")
plt.grid()
plt.xlabel('WHOR')
plt.ylabel('New cases-deaths')
plt.plot(df['WHOR'],df['New cases'],'k-->')
plt.plot(df['WHOR'],df['New deaths'],'r:')
plt.xticks(rotation=90)
plt.show()
```

```
plt.figure(figsize=(10,5))
plt.title("JINENDRA_MEHTA_180570107070")
plt.grid()
plt.xlabel('WHOR')
plt.ylabel('New cases-deaths')
plt.plot(df['WHOR'],df['New cases'],'k-->')
plt.plot(df['WHOR'],df['New deaths'],'r:o')
plt.xticks(rotation=90)
plt.show()
```

```
plt.figure(figsize=(10,5))
plt.title("JINENDRA_MEHTA_180570107070")
plt.grid()
plt.xlabel('WHOR')
plt.ylabel('New cases-deaths')
plt.plot(df['WHOR'],df['New cases'],'k-->')
plt.plot(df['WHOR'],df['New deaths'],'r:o')
pick_cases=max(df['New cases'])
plt.annotate(xy=[df[df['New cases']==df['New cases'].max()].WHOR,pick_cases],s='NewCasesPick')
plt.xticks(rotation=90)
plt.show()
```

```
plt.figure(figsize=(10,5))
plt.title("JINENDRA_MEHTA_180570107070")
plt.grid()
plt.xlabel('WHOR')
plt.ylabel('New cases-deaths')
plt.plot(df['WHOR'],df['New cases'],'k-->')
plt.plot(df['WHOR'],df['New deaths'],'r:o')
pick_cases=max(df['New cases'])
plt.annotate(xy=[df[df['New cases']==df['New cases'].max()].WHOR,pick_cases],s='NewCasesPick')
pick_deaths=max(df['New deaths'])
plt.annotate(xy=[df[df['New deaths']==df['New deaths'].max()].WHOR,pick_deaths],s='NewDeathsPick')
plt.xticks(rotation=90)
plt.show()
```
